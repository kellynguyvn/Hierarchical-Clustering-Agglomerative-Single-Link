```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Specific data points
X = np.array([[2, 2],
              [2, 6],
              [3, 7],
              [5, 2],
              [5, 5],
              [5, 8],
              [6, 6],
              [8, 4],
              [7, 3],
              [10, 6],
              [12, 8]])
```

```python
# Creating Agglomerative Clustering models
single_link = AgglomerativeClustering(n_clusters=3, linkage='single')
complete_link = AgglomerativeClustering(n_clusters=3, linkage='complete')
average_link = AgglomerativeClustering(n_clusters=3, linkage='average')
```

```python
# Fitting the models
single_link.fit(X)
complete_link.fit(X)
average_link.fit(X)
```

```python
# Plotting Single-Link clustering
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=single_link.labels_, cmap='rainbow')
plt.title('Single-Link Clustering')
plt.show()
```

```python
# Plotting Complete-Link clustering
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=complete_link.labels_, cmap='rainbow')
plt.title('Complete-Link Clustering')
plt.show()
```

```python
# Plotting Average-Link clustering
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=average_link.labels_, cmap='rainbow')
plt.title('Average-Link Clustering')
plt.show()
```
